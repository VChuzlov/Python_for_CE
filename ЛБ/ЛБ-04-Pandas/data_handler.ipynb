```python
import pandas as pd
```

```python
df = pd.read_csv('db.csv')
```

```python
for i, val in enumerate(df['class']):
    if val == 'N':
        df.loc[i, 'class'] = 'Ni'
```

```python
for i, val in enumerate(df['class']):
    if val == 'n':
        df.loc[i, 'class'] = 'N'
```

```python
df
```

```python
names_map = {}
sub = 'Value'

for name in df.columns:
    if name.endswith(sub):
        names_map[name] = name.replace(sub, '')

df = df.rename(columns=names_map)
df.to_excel('compounds_data.xlsx', index=False)
```
